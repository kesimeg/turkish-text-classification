```python
import fasttext
from datasets import load_dataset
import os
import requests
from tqdm import tqdm

dataset = load_dataset("stefan-it/offenseval2020_tr")

train_data = dataset['train']
valid_data = dataset['validation']
test_data = dataset['test']

train_data.to_csv("train.txt") #fast text wants a txt file as input
valid_data.to_csv("valid.txt")
test_data.to_csv("test.txt")
```

```python
!wget https://dl.fbaipublicfiles.com/fasttext/vectors-crawl/cc.tr.300.vec.gz #download turkish word vectors
```

```python
!gzip -d cc.tr.300.vec.gz
```

```python
classifier = fasttext.train_supervised(
    input='train.txt',
    pretrained_vectors='cc.tr.300.vec', 
    epoch=1,
    wordNgrams=2,
    dim=300,
    loss='softmax',
    verbose=2
)
```

```python
result = classifier.test('train.txt')
print(f"Train accuracy: {result[1] * 100:.2f}%")
result = classifier.test('valid.txt')
print(f"Validation accuracy: {result[1] * 100:.2f}%")
result = classifier.test('test.txt')
print(f"Test accuracy: {result[1] * 100:.2f}%")
```
